```python
#Задача на поиск самой длинной CG-последовательности и выводом её ID и процентов
filename = 'rosalind_gc.txt'       #Имя файла

with open("C:\\Users\\Полина\\Downloads\\rosalind_gc.txt", 'r') as file:
    lines = file.readlines()

sequences = {}                     #Создаём словарь для хранения ID и последовательностей
current_id = ''

for line in lines:
    line = line.strip()
    if line.startswith('>'):        #Начинаем новая запись, сохраняем новый ID без '>'
        current_id = line[1:]
        sequences[current_id] = ''
    else:                           #Продолжаем добавлять нуклеотиды к текущей последовательности
        sequences[current_id] += line

def gc_content(dna):                 #Расчёта GC-содержания
    gc_count = 0
    for nucleotide in dna:
        if nucleotide == 'G' or nucleotide == 'C':
            gc_count += 1
    return (gc_count / len(dna)) * 100

max_id = ''                          # Поиск последовательности с максимальным GC-содержанием
max_gc = 0.0

for seq_id, dna_seq in sequences.items():
    current_gc = gc_content(dna_seq)
    if current_gc > max_gc:
        max_gc = current_gc
        max_id = seq_id
        
print(max_id)
print(f"{max_gc:.6f}")
```

```python
#Задача на поиск число позиций, в которых две строки одинаковой длины отличаются друг от друга
filename = 'rosalind_hamm.txt'
with open("C:\\Users\\Полина\\Downloads\\rosalind_hamm.txt", 'r') as file:  #Считываем две строки из файла
    dna1 = file.readline().strip()
    dna2 = file.readline().strip()

if len(dna1) != len(dna2):         #Проверяем, что длины строк равны
    print("Ошибка: строки разной длины")
    exit(1)

def hamming_distance(s1, s2):       #Функция для подсчёта расстояния Хэмминга
    count = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            count += 1
    return count

distance = hamming_distance(dna1, dna2) #Вычисляем расстояние
print(distance)
```

```python
#Задача на 1 закон Менделя
with open("C:\\Users\\Полина\\Downloads\\rosalind_iprb(9).txt", 'r') as file:
    data = file.read().split()
    k = int(data[0])  #Гомозигота АА
    m = int(data[1])  #Гетерозигота Аа
    n = int(data[2])  #Гомозигота аа

total = k + m + n     #Общее количество особей в популяции

# Вероятность выбрать две конкретные особи (без возвращения)
# Вероятность выбрать первую особь типа X: X/total
# Вероятность выбрать вторую особь типа Y: Y/(total-1) 
# (если X ≠ Y) или (X-1)/(total-1) (если X = Y)

# Расчет вероятности рецессивного потомства (aa)
recessive_prob = 0.0

# 1. Оба родителя гомозиготные рецессивные (aa × aa)
# Вероятность выбрать двух рецессивных: [n/total] * [(n-1)/(total-1)]
# Вероятность рецессивного потомства: 1.0
if n >= 2:
    recessive_prob += (n / total) * ((n - 1) / (total - 1)) * 1.0

# 2. Оба родителя гетерозиготные (Aa × Aa)
# Вероятность выбрать двух гетерозиготных: [m/total] * [(m-1)/(total-1)]
# Вероятность рецессивного потомства: 0.25 (закон Менделя)
if m >= 2:
    recessive_prob += (m / total) * ((m - 1) / (total - 1)) * 0.25

# 3. Один родитель гетерозиготный, другой - рецессивный (Aa × aa)
# Вероятность такой пары: 
# [m/total * n/(total-1)] + [n/total * m/(total-1)] = 2 * (m/total) * (n/(total-1))
# Вероятность рецессивного потомства: 0.5
if m >= 1 and n >= 1:
    recessive_prob += 2 * (m / total) * (n / (total - 1)) * 0.5
#остальные вероятности не учитываем
dominant_prob = 1 - recessive_prob     # Итоговая вероятность доминантного потомства

print("{:.5f}".format(dominant_prob))  #Округляем до 5 знака после запятой
```

```python
#Задача на поиск мажорного элемента(который встречается строго больше n/2 раз в каждом массиве)
def find_majority_element(arr): #Функция для поиска мажоритарного элемента в массиве, возвращает элемент, если он встречается > n/2 раз, иначе -1.

    candidate = None      # Первый проход: находим кандидата
    count = 0
    
    for num in arr:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1

    if candidate is not None:        # Второй проход: проверяем, действительно ли кандидат мажоритарный
        actual_count = 0
        for num in arr:
            if num == candidate:
                actual_count += 1
        
        if actual_count > len(arr) // 2:
            return candidate
    
    return -1

with open("C:\\Users\\Полина\\Downloads\\rosalind_maj.txt", 'r') as file:       #Чтение входных данных из файла
    data = file.readlines()

first_line = data[0].split()    #Первая строка: количество массивов и размер каждого массива
k = int(first_line[0])  # количество массивов
n = int(first_line[1])  # размер каждого массива

results = []

for i in range(1, k + 1): #Обрабатываем каждый массив
    arr = list(map(int, data[i].split())) #Читаем массив чисел
    result = find_majority_element(arr) # Ищем мажоритарный элемент
    results.append(str(result))

print(' '.join(results))       #Выводим результаты через пробел
```

```python
#задача на слияние двух отсортированных массивов
def merge_sorted_arrays(arr1, arr2): #Сливаем 2 отсортированных массива
    result = [] 
    i = j = 0

    while i < len(arr1) and j < len(arr2): # Проходим по обоим массивам, пока не дойдем до конца одного из них
        if arr1[i] <= arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1

    while i < len(arr1):                 #Добавляем оставшиеся элементы из первого массива (если есть)
        result.append(arr1[i])
        i += 1

    while j < len(arr2):                 #Добавляем оставшиеся элементы из второго массива (если есть)
        result.append(arr2[j]) 
        j += 1
    
    return result

with open("C:\\Users\\Полина\\Downloads\\rosalind_mer.txt", 'r') as file: #Чтение входных данных из файла
    data = file.readlines()
    
    n = int(data[0].strip())                                      #Читаем размер первого массива и сам массив
    arr1 = list(map(int, data[1].split()))

    m = int(data[2].strip())                                      #Читаем размер второго массива и сам массив
    arr2 = list(map(int, data[3].split()))

merged_array = merge_sorted_arrays(arr1, arr2)                    #Сливаем два отсортированных массива
output = ' '.join(map(str, merged_array))                         #Форматируем вывод: преобразуем числа в строки и соединяем пробелами
print(output)
```

```python
#Задача на вычисление суммы степеней соседей каждой вершины
with open("C:\\Users\\Полина\\Downloads\\rosalind_ddeg.txt", 'r') as file:
    lines = file.readlines() # Читаем все строки файла

first_line = lines[0].split()    # Первая строка содержит количество вершин и ребер
n = int(first_line[0])  # количество вершин
m = int(first_line[1])  # количество ребер

graph = [[] for _ in range(n + 1)] # Создаем пустой граф (список смежности), используем n+1 элементов, чтобы удобно работать с вершинами от 1 до n

for i in range(1, m + 1):          #Читаем все ребра и заполняем граф
    edge_line = lines[i].split()   
    u = int(edge_line[0])  #первая вершина ребра
    v = int(edge_line[1])  #вторая вершина ребра

    graph[u].append(v) # Добавляем связь в обе стороны (граф неориентированный)
    graph[v].append(u)

degrees = [0] * (n + 1) # Вычисляем степень каждой вершины, степень вершины - сколько у нее соседей
for i in range(1, n + 1):
    degrees[i] = len(graph[i])

result = []             #Теперь для каждой вершины считаем сумму степеней ее соседей
for vertex in range(1, n + 1):
    total_neighbor_degree = 0

    for neighbor in graph[vertex]:  #Проходим по всем соседям текущей вершины
        total_neighbor_degree += degrees[neighbor] #Прибавляем степень соседа

    result.append(str(total_neighbor_degree)) #Добавляем результат для этой вершины

print(' '.join(result)) #Выводим результат через пробел
```
